```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
#import modul
```

```python
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
#sekumpulan bilangan pecahan/ float
```

```python
plt.scatter(x, y, label='Data 1', color='r')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')

plt.legend()

plt.show()
#data point ditandai oleh X  dan Y
#paramater label berkolerasi dengan data legend
```

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
#setiap variabel menampung bilangan
```

```python
plt.scatter(x1, y1, label='Data 1', color='r')
plt.scatter(x2, y2, label='Data 2', color='b')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Multiple Scatter Plot')

plt.legend()

plt.show()
#plotting data pertama dan kedua
```

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x1, y1,
           color='cyan',
           linewidths=2,
           marker='s',
           edgecolor='red',
           s=400)

plt.scatter(x2, y2,
           color='yellow',
           linewidths=1,
           marker='^',
           edgecolor='blue',
           s=200)

plt.show()
#plotting marker 1 warna cyan
#plotting marker 2 color yellow
```
